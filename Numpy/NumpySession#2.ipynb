CREACIÓN DE CAFETERIA DIGITAL

```python
#Creación de la matriz de la cafetería
#[cafés vendidos, croissants vendidos]
ventas = np.array([
    [10,5],#lunes
    [15,8],#martes
    [12,6],#miercoles
    [20,12],#jueves
    [25,15],#viernes
])
print("Nuestra tabla de ventas:")
print(ventas)
```

```python
#Mostar la primera fila
ventas[0]
```

```python
#Mostar la segunda columna
croissants_vendidos = ventas[:,1]
print(croissants_vendidos)
```

```python
ventas_inicio_semana = ventas [0:2, :]
print("Solo Lunes y Martes")
print(ventas_inicio_semana)
```

Array Broadcasting

```python
a = np.array([1, 2, 3])
b = 2

#SUMA
#Suma a cada elemento del array 2
result = a + b
print(result)
```

```python
#MATRIZ
#Le suma la dimension del arr a cada dimension de la matriz
arr = np.array([1, 2, 3])
matriz =np.array([
    [1, 2, 3],
    [4, 5, 6]])

result = arr + matriz
print(result)
```

MEGAFONO MATEMÁTICO

```python
#Creación de la matriz de la cafetería
#[cafés vendidos, croissants vendidos]
ventas = np.array([
    [10,5],#lunes
    [15,8],#martes
    [12,6],#miercoles
    [20,12],#jueves
    [25,15],#viernes
])
print("Nuestra tabla de ventas:")
print(ventas)
```

```python
#Sumar .50 a cada precio
ventas_ajuste = ventas + 0.50
print("Precios ajustados por 0.50")
print(ventas_ajuste)
```

Sistemas Lineales


```python
#Código Sistemas Lineales

#A Array
A = np.array([
    [2,1],#lunes
    [1,1],#martes
])

#B De lo que se paga
B = np.array([5, 3])

result = np.linalg.solve(A, B)
print(f"Precio Café: {result[0]}")
print(f"Precio Croissant: {result[1]}")
```
