ACCEDER A ELEMENTOS (indexing)

```python
#acceder a los elementos (igual en las listas)
#Los indices empiezan en 0
a = np.array([1, 2, 3, 4, 5])
a
```

```python
print(a[0]) #Primera posicion
```

```python
print(a[4]) #Ultima posicion

print(a[-1]) #Otra manera de obtener Ultima posicion
```

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
a
```

```python
#Elemento especifico de la fila1
print(a[0, 0])
```
